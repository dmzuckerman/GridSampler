```python
# prompt: calculate a similarity score between every pair of names based on the number of letters in common and show the results
#@title Pairwise analysis of names

from collections import Counter

first_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']

def letter_similarity(name1, name2):
    """Calculates a similarity score based on common letters."""
    counter1 = Counter(name1.lower())
    counter2 = Counter(name2.lower())
    common_letters = sum((counter1 & counter2).values())
    # A simple scoring mechanism: number of common letters
    return common_letters

print("Similarity Scores:")
for i in range(len(first_names)):
    for j in range(i + 1, len(first_names)):
        name1 = first_names[i]
        name2 = first_names[j]
        score = letter_similarity(name1, name2)
        print(f"  {name1} and {name2}: {score}")
```

```python
#@title Make synthetic data: lists of x and y values (complicated - ignore this)

import matplotlib.pyplot as plt
import numpy as np

# Generate 100 random points in two dimensions, clustered in three regions
np.random.seed(0)  # for reproducibility

n_points = 100
n_clusters = 3

x = []
y = []

for i in range(n_clusters):
  # Generate points for each cluster
  cluster_x = np.random.normal(loc=np.random.uniform(low=2, high=8), scale=1, size=n_points // n_clusters)
  cluster_y = np.random.normal(loc=i * 2 + 2, scale=0.75, size=n_points // n_clusters)
  x.extend(cluster_x)
  y.extend(cluster_y)

# Add a few outliers
x.extend(np.random.uniform(low=0, high=9, size=n_points % n_clusters))
y.extend(np.random.uniform(low=0, high=6, size=n_points % n_clusters))

plt.scatter(x, y)
plt.xlabel("Gene A expression")
plt.ylabel("Gene B expression")
plt.show()
```

```python
#@title FIRST STEP OF K-MEANS CLUSTEERING: Set initial cluster centers randomly
np.random.seed(1)  # for reproducibility

n_clusters = 3

x_centers = []
y_centers = []

for i in range(n_clusters):
  # Generate points for each cluster
  x_centers.append(np.random.uniform(low=2, high=8))
  y_centers.append(np.random.uniform(low=2, high=8))

plt.scatter(x, y)
plt.scatter(x_centers, y_centers, color='red', label = 'Initial cluster centers')
plt.xlabel("Gene A expression")
plt.ylabel("Gene B expression")
plt.legend() # Show the legend to identify clusters and centers
plt.show()
```

```python
#@title NEXT: FIND POINTS ASSOCIATED TO CLUSTER CENTERS

cluster0 = []
cluster1 = []
cluster2 = []

for i in range(n_points):
  min = 10000 # a number bigger than any possible distance
  for j in range(n_clusters):
    distance = np.sqrt((x[i] - x_centers[j])**2 + (y[i] - y_centers[j])**2)
    if distance < min:
      min = distance
      cluster = j
      # print('point ',i, '  cluster',j)
  if cluster == 0:
    cluster0.append([x[i].item(),y[i].item()])
  elif cluster == 1:
    cluster1.append([x[i].item(),y[i].item()])
  elif cluster == 2:
    cluster2.append([x[i].item(),y[i].item()])

#print('cluster0', cluster0)
#print('cluster1', cluster1)
#print('cluster2', cluster2)
```

```python
#@title Plot clustering results (complicated - plotting details)

import matplotlib.pyplot as plt
import numpy as np
cluster0 = np.array(cluster0)
cluster1 = np.array(cluster1)
cluster2 = np.array(cluster2)

plt.figure(figsize=(8, 6)) # Optional: Set the figure size

# Plot each cluster with a different color
if len(cluster0) > 0:
    plt.scatter(cluster0[:, 0], cluster0[:, 1], color='blue', label='Cluster 0')
if len(cluster1) > 0:
    plt.scatter(cluster1[:, 0], cluster1[:, 1], color='green', label='Cluster 1')
if len(cluster2) > 0:
    plt.scatter(cluster2[:, 0], cluster2[:, 1], color='purple', label='Cluster 2')

# Plot cluster centers with a larger symbol (e.g., 'X')
plt.scatter(x_centers[0], y_centers[0], color='blue', marker='X', s=150, label='Center 0')
plt.scatter(x_centers[1], y_centers[1], color='green', marker='X', s=150, label='Center 1')
plt.scatter(x_centers[2], y_centers[2], color='purple', marker='X', s=150, label='Center 2')


plt.xlabel("Gene A expression")
plt.ylabel("Gene B expression")
plt.title("K-Means Clustering")
plt.legend() # Show the legend to identify clusters and centers
plt.grid(True) # Optional: Add a grid
plt.show()
```

```python
#@title SECOND STEP OF K-MEANS: Calculate new cluster centers as averages in each cluster

x_centers_new = []
y_centers_new = []

# Process cluster 0 (first cluster)
xsum = 0
ysum = 0
for i in range(len(cluster0)):
    xsum += cluster0[i][0]
    ysum += cluster0[i][1]
x_centers_new.append( xsum / len(cluster0) )
y_centers_new.append( ysum / len(cluster0) )

```

```python
#@title Compute last two new cluster centers (same code as for first center)

# Process cluster 1 (second cluster)
xsum = 0
ysum = 0
for i in range(len(cluster1)):
    xsum += cluster1[i][0]
    ysum += cluster1[i][1]
x_centers_new.append( xsum / len(cluster1) )
y_centers_new.append( ysum / len(cluster1) )

# Last cluster
xsum = 0
ysum = 0
for i in range(len(cluster2)):
    xsum += cluster2[i][0]
    ysum += cluster2[i][1]
x_centers_new.append( xsum / len(cluster2) )
y_centers_new.append( ysum / len(cluster2) )

```

```python
#@title PLOT CLUSTERS SHOWING NEW CLUSTER CENTERS (complicated - plotting details)

plt.figure(figsize=(8, 6)) # Optional: Set the figure size

# Plot each cluster with a different color
if len(cluster0) > 0:
    plt.scatter(cluster0[:, 0], cluster0[:, 1], color='blue', label='Cluster 0')
if len(cluster1) > 0:
    plt.scatter(cluster1[:, 0], cluster1[:, 1], color='green', label='Cluster 1')
if len(cluster2) > 0:
    plt.scatter(cluster2[:, 0], cluster2[:, 1], color='purple', label='Cluster 2')

# Plot cluster centers with a larger symbol (e.g., 'X')
plt.scatter(x_centers[0], y_centers[0], color='blue', marker='X', s=150, label='Center 0')
plt.scatter(x_centers[1], y_centers[1], color='green', marker='X', s=150, label='Center 1')
plt.scatter(x_centers[2], y_centers[2], color='purple', marker='X', s=150, label='Center 2')

# Plot NEW cluster centers with a larger symbol (e.g., 'X')
plt.scatter(x_centers_new[0], y_centers_new[0], color='blue', marker='o', s=150, label='New Center 0')
plt.scatter(x_centers_new[1], y_centers_new[1], color='green', marker='o', s=150, label='New Center 1')
plt.scatter(x_centers_new[2], y_centers_new[2], color='purple', marker='o', s=150, label='New Center 2')

plt.xlabel("Gene A expression")
plt.ylabel("Gene B expression")
plt.title("K-Means Clustering")
plt.legend() # Show the legend to identify clusters and centers
plt.grid(True) # Optional: Add a grid
plt.show()
```

```python
#@title FIND POINTS ASSOCIATED TO NEW CLUSTER CENTERS (repeats prior code with new centers)

cluster0 = []
cluster1 = []
cluster2 = []

for i in range(n_points):
  min = 10000 # a number bigger than any possible distance
  for j in range(n_clusters):
    distance = np.sqrt((x[i] - x_centers_new[j])**2 + (y[i] - y_centers_new[j])**2)
    if distance < min:
      min = distance
      cluster = j
      # print('point ',i, '  cluster',j)
  if cluster == 0:
    cluster0.append([x[i].item(),y[i].item()])
  elif cluster == 1:
    cluster1.append([x[i].item(),y[i].item()])
  elif cluster == 2:
    cluster2.append([x[i].item(),y[i].item()])

cluster0 = np.array(cluster0)
cluster1 = np.array(cluster1)
cluster2 = np.array(cluster2)

plt.figure(figsize=(8, 6)) # Optional: Set the figure size

# Plot each cluster with a different color
if len(cluster0) > 0:
    plt.scatter(cluster0[:, 0], cluster0[:, 1], color='blue', label='Cluster 0')
if len(cluster1) > 0:
    plt.scatter(cluster1[:, 0], cluster1[:, 1], color='green', label='Cluster 1')
if len(cluster2) > 0:
    plt.scatter(cluster2[:, 0], cluster2[:, 1], color='purple', label='Cluster 2')

# Plot cluster centers with a larger symbol (e.g., 'X')
plt.scatter(x_centers[0], y_centers[0], color='blue', marker='X', s=150, label='Center 0')
plt.scatter(x_centers[1], y_centers[1], color='green', marker='X', s=150, label='Center 1')
plt.scatter(x_centers[2], y_centers[2], color='purple', marker='X', s=150, label='Center 2')

# Plot NEW cluster centers with a larger symbol (e.g., 'X')
plt.scatter(x_centers_new[0], y_centers_new[0], color='blue', marker='o', s=150, label='New Center 0')
plt.scatter(x_centers_new[1], y_centers_new[1], color='green', marker='o', s=150, label='New Center 1')
plt.scatter(x_centers_new[2], y_centers_new[2], color='purple', marker='o', s=150, label='New Center 2')

plt.xlabel("Gene A expression")
plt.ylabel("Gene B expression")
plt.title("K-Means Clustering")
plt.legend() # Show the legend to identify clusters and centers
plt.grid(True) # Optional: Add a grid
plt.show()
```
